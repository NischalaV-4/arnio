# 🚀 Arnio: High-Performance Data Quality Pipeline Tutorial

Welcome to the beginner's guide to **Arnio**! This notebook demonstrates how to build an automated, high-speed data cleaning pipeline to handle messy datasets seamlessly.

```python
import pandas as pd

# 1. Setup sample messy dataset
messy_data = {
    "student_id": [101, 102, 102, 103, 104],
    "name": ["  Nishtha ", "Anish", "Anish", "  Siddharth", "Rahul "],
    "city": ["delhi", "mumbai", "mumbai", "DELHI", "bangalore"]
}

df = pd.DataFrame(messy_data)
print("--- Raw Input Data Matrix ---")
print(df)
print("-" * 30)

# 2. Executing Data Transformation Architecture
def arnio_clean_pipeline(data_frame):
    data_frame["name"] = data_frame["name"].str.strip()
    data_frame["city"] = data_frame["city"].str.lower()
    cleaned_df = data_frame.drop_duplicates()
    return cleaned_df

processed_df = arnio_clean_pipeline(df)
print("\n--- Processed Pristine Dataset ---")
print(processed_df)
```
